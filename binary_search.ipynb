```python
"""
   come up with posible example on input and output and clearly define the edge cases,
   put them in a list then loop the test the list using one function
"""

tests = []
# query appears at the middle
tests.append({
    "input": {
        "cards": [9, 7, 5, 3, 2],
        "query": 5
    },
    "output": 2  # index of the query
})
# query appears in the first element
tests.append({
    "input": {
        "cards": [2, 4, 5, 8],
        "query": 2
    },
    "output": 0
})
# query appears in the last element
tests.append({
    "input": {
        "cards": [3, 2, -4, -6],
        "query": -6
    },
    "output": 3
})
# card has only one element which is query
tests.append({
    "input": {
        "cards": [5],
        "query": 5
    },
    "output": 0
})
# assume fn return -1 when there's no query in cards or cards is empty
tests.append({
    "input": {
        "cards": [2, 5, 7, 2],
        "query": 9
    },
    "output": -1
})
# cards is empty
tests.append({
    "input": {
        "cards": [],
        "query": 3
    },
    "output": -1
})
tests.append({
    "input": {
        "cards": [8, 8, 6, 6, 6, 4, 5, 7],
        "query": 6
    },
    "output": 2
})

print(tests)
```

```python
def locate_card(cards,query):
    # create postion for index
    position = 0
    # create a loop
    while position < len(cards):
        # check index value is equal to query
        if cards[position] == query:
            # answer found and exit

            return position
        # increament index
        position += 1
        
        if len(cards) == position:
            # check the end of the loop
            return -1
        
#refactor of the above function
def locate_card_for(cards,query):
    position = 0
    for i in range(len(cards)):
        if cards[position] == query:
            return position
        else:
            position += 1
        
```

```python
demo = {
    "input": {
        "cards": [9, 7, 5, 3, 2],
        "query": 5
    },
    "output": 2  # index of the query
}
results = locate_card(demo["input"]["cards"], demo["input"]["query"])
print(results == demo["output"])
res = locate_card_for(demo["input"]["cards"], demo["input"]["query"])
print(res == demo["output"])
```

```python
# testing our function agains all posible scenarios
# evaluate tests
from jovian.pythondsa import evaluate_test_cases
evaluate_test_cases(locate_card,tests)
```

```python
def binary_locate_card(cards,query):
    low, high = 0, len(cards) - 1
    while low <= high:
        mid = low + high // 2
        mid_number = cards[mid]
        print(f"The low is {low} The highest is {high} The mid number is {mid_number}")

        if mid_number == query:
            return mid
        elif mid_number < query:
            high = mid - 1
        elif mid_number > query:
            low = mid + 1
    return -1
        
```

```python
# binary search refactor to accomodate dublicate numbers in a cards list
def test_location(cards, query, mid):
    mid_number = cards[mid]
    if mid_number == query:
        if mid-1 >= 0 and cards[mid - 1] == query:
            return "left"
        else:
            return "found"
    elif mid_number < query:
        return "left"
    else:
        return "right"


def locate_card(cards, query):
    low, high = 0, len(cards) - 1
    while low <= high:
        mid = low + high // 2
        results = test_location(cards, query, mid)

        if results == "found":
            return mid
        elif results == "left":
            high = mid - 1
        elif results == "right":
            low = mid + 1
    return -1
```

```python
"""Example"""
# check position
def count_rotations_linear(nums):
    position = 0

    while position < len(position):
        if position > 0 and nums[position] < nums[position - 1]:
            return position
        position += 1
    return 0
```
